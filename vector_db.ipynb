```python
import json
from pprint import pprint

from langchain_community.document_loaders import JSONLoader, DirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_core.prompts import ChatPromptTemplate
```

# 폴더 내의 모든 JSON 파일 로드

```python
loader = DirectoryLoader(path='/content/drive/MyDrive/이어드림/랭체인/data/', glob='*.json')
data = loader.load()
len(data)
```

```python
docs = []

for i in range(len(data)):
    loader = JSONLoader(
        file_path=data[i].metadata['source'],
        jq_schema=".[].Content",
        text_content=False,
    )
    docs.extend(loader.load())
    print(f"문서의 수: {len(docs)}")
```

```python
docs[0].page_content
```

```python
for i in range(len(docs)):
    docs[i].page_content = docs[i].page_content.encode('utf-8').decode('unicode-escape')

print(docs[0].page_content)
```

```python
pprint(docs[0])
pprint(docs[-1])
```

# 문서 분할하여 VectorDB 구축

```python
# docs 리스트를 10등분하여 각각 docs1, docs2, ..., docs10에 할당
docs_split = []
split_size = len(docs) // 10

for i in range(10):
    start_idx = i * split_size
    end_idx = (i + 1) * split_size if i < 9 else len(docs)  # 마지막 부분은 나머지를 포함
    docs_split.append(docs[start_idx:end_idx])

# 각 분할된 리스트에 접근
docs1, docs2, docs3, docs4, docs5, docs6, docs7, docs8, docs9, docs10 = docs_split
len(docs1)
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)

split_docs = []

# docs_split에 있는 각 문서를 text_splitter로 분할하여 split_docs에 저장
for docs in docs_split:
    split_docs.append(text_splitter.split_documents(docs))

len(split_docs)
```

```python
len(split_docs[-1])
```

```python
split_docs[0][1].page_content
```

```python
embeddings = HuggingFaceEmbeddings(
    model_name='BAAI/bge-m3',
    model_kwargs={'device':'cuda'},
    encode_kwargs={'normalize_embeddings':True},
)
```

```python
def save_vectorstore(docs, folder_path, index_name, embeddings):
    vectorstore = FAISS.from_documents(docs, embeddings)
    vectorstore.save_local(folder_path=folder_path, index_name=index_name)

for i in range(10):
    folder_path = f'faiss_db{i+1}'
    index_name = f'faiss_index{i+1}'
    save_vectorstore(split_docs[i], folder_path, index_name, embeddings)
```

# vectorDB merge
- https://wikidocs.net/234014

```python
# 첫 번째 벡터스토어 로드
db = FAISS.load_local(
    folder_path="/content/faiss_db1",
    index_name="faiss_index1",
    embeddings=embeddings,
    allow_dangerous_deserialization=True,
)

# 2번부터 10번까지 벡터스토어를 순차적으로 로드하고 병합
for i in range(2, 11):
    db_to_merge = FAISS.load_local(
        folder_path=f"/content/faiss_db{i}",
        index_name=f"faiss_index{i}",
        embeddings=embeddings,
        allow_dangerous_deserialization=True,
    )
    db.merge_from(db_to_merge)
```
